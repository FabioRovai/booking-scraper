```python
#from bs4 import BeautifulSoup
#soup = BeautifulSoup(str_dictionary, 'html.parser')
#print(soup.prettify())
text_file = open("/content/sample.txt", "r")
lines = text_file.readlines()
#find word class="class="bui-card__title"
lines= str(lines)[str(lines).find('class="bui-card__title'):]
words = str(lines).split('class="bui-card__title" itemprop="name">\\\\n')

baz=[i.split('\\') for i in words]
#name
n=1
while n <= (len(baz)-1):
    try:
      print(baz[n][0])
      file = open("names.txt", "a")
      file.write(baz[n][0] + "\n") 
      n+=1
    except IndexError:
      print ('NaN')
      file = open("names.txt", "a")
      file.write('NaN' + "\n")
      n+=1
```

```python
#stars
#baz[1][14] #debug index
import re


n=1
while n <= (len(baz)-1):
  try:
    stars=list(filter(lambda a: 'stars' in a, baz[n]))
    
    stars= stars[1].split('>')
    print(re.sub('[^0-9]', '', str(stars)))
    file = open("stars.txt", "a")
    file.write(re.sub('[^0-9]', '', str(stars)) + "\n")    
    n+=1
  except IndexError:
    print ('NaN')
    file = open("stars.txt", "a")
    file.write('NaN' + "\n")
    n+=1
    
```

```python
#loction
#baz[10][40]
n=1
while n <= (len(baz)-1):
  try:
    hotel=list(filter(lambda a: ', Rome' in a, baz[n]))
    
    loction= hotel[0].split('>')
    print(loction)
    file = open("location.txt", "a")
    file.write(loction[0] + "\n")
    n+=1
  except IndexError:
    print ('NaN')
    file = open("location.txt", "a")
    file.write('NaN' + "\n")
    n+=1
```

```python
#top review
#baz[1][56] #hotel-card__text_review

n=1
while n <= (len(baz)-1):
  try:
    hotel=list(filter(lambda a: 'hotel-card__text_review' in a, baz[n]))
    
    loction= hotel[0].split('>')
    
    print([loction[1].strip('</span') for i in loction])
    file = open("review.txt", "a")
    file.write(loction[1].strip('</span') + "\n")
    n+=1
  except IndexError:
    print ('NaN')
    file = open("review.txt", "a")
    file.write('NaN' + "\n")
    n+=1
```

```python
#score
#baz[1][96-16]

n=1
while n <= (len(baz)-1):
  try:
    hotel=list(filter(lambda a: 'Scored ' in a, baz[n]))
    
    loction= hotel[0].split('aria-label="')
    loction=loction[1].strip('Scored')
    #print(loction[1])
    file = open("Scored.txt", "a")
    file.write(loction.strip('">') + "\n")
    n+=1
  except IndexError:
    print ('NaN')
    file = open("Scored.txt", "a")
    file.write('NaN' + "\n")
    n+=1
```

```python
#print(baz[1][112])
#print(baz[1][114])
#£
n=1
while n <= (len(baz)-1):
  try:
    hotel=list(filter(lambda a: '£' in a, baz[n]))
    
    hotel= hotel[0].split('£')

    hoteel=hotel[1].strip('per night')
    print(hoteel.strip('</div>'))
    file = open("£.txt", "a")
    file.write(hoteel.strip('</div>') + "\n")
    n+=1
  except IndexError:
    print ('NaN')
    file = open("£.txt", "a")
    file.write('NaN' + "\n")
    n+=1
```

```python
text_file = open("/content/Scored.txt", "r")
SDG = text_file.readlines()
SDG=pd.DataFrame(SDG)
SDG.to_csv('scored.csv')
scored=pd.read_csv('scored.csv')
#rename
scored.rename(columns={ scored.columns[1]: "scored" }, inplace = True)

scored.iloc[:,1]
```

```python
text_file = open("/content/£.txt", "r")
SDG = text_file.readlines()
SDG=pd.DataFrame(SDG)
SDG.to_csv('£.csv')
pound=pd.read_csv('£.csv')
#rename
pound.rename(columns={ pound.columns[1]: "£" }, inplace = True)
pound['£'] = pound['£'].str.strip('\n')

pound.iloc[:,1]
```

```python
text_file = open("/content/location.txt", "r")
SDG = text_file.readlines()
SDG=pd.DataFrame(SDG)
SDG.to_csv('location.csv')
location=pd.read_csv('location.csv')
#rename
location.rename(columns={ location.columns[1]: "location" }, inplace = True)
location['location'] = location['location'].str.strip('\n')
for col in location['location']:
    location['location'] = location['location'].apply(lambda x : x[1:] if x.startswith("n") else x)
location.iloc[:,1]
```

```python
text_file = open("/content/stars.txt", "r")
SDG = text_file.readlines()
SDG=pd.DataFrame(SDG)
SDG.to_csv('stars.csv')
stars=pd.read_csv('stars.csv')
#rename
stars.rename(columns={ stars.columns[1]: "stars" }, inplace = True)
stars['stars'] = stars['stars'].str.strip('\n')
stars.iloc[:,1]
```

```python
text_file = open("/content/names.txt", "r")
SDG = text_file.readlines()
SDG=pd.DataFrame(SDG)
SDG.to_csv('names.csv')
names=pd.read_csv('names.csv')
#rename
names.rename(columns={ names.columns[1]: "names" }, inplace = True)
names['names'] = names['names'].str.strip('\n')
names.iloc[:,1]
```
